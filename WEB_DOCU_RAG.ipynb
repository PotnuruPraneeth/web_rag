```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
from bs4 import BeautifulSoup #helps parse and search HTML.
import requests#The requests library is a powerful and user-friendly tool used to send HTTP requests in Python — basically, it helps your Python code connect to websites and get data.
```

```python
req=requests.get('https://directai.blog/')## Sending a GET request to the website
```

```python
scoup=BeautifulSoup(req.text,'html.parser')#'html.parser': Python’s built-in HTML parser.
```

```python
print(scoup)
```

```python
links = scoup.find_all('a')
print(links) # <a href="https://www.example.com">Click here</a>
```

```python
urls = [link.get('href') for link in links if link.get('href') is not None]
print(urls)
```

```python
url=[]
for i in urls:
    if i.startswith('https'):
        url.append(i)
print(url)
```

```python
loader = WebBaseLoader(url[0:10]) ## Initialize WebBaseLoader with the list of URLs
print(loader)
```

```python
documents=loader.load()
print(documents)
```

```python
for doc in documents:
    print(doc.metadata['source'])
    print(doc.metadata['title'])
    print(doc.page_content)
```

```python
print(documents[7])# means 7th link
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter # by this you can split any kind of the format data
```

```python
spliter=RecursiveCharacterTextSplitter( chunk_size=50,
        chunk_overlap=10,
        length_function=len,
        is_separator_regex=False,)
chunck=spliter.split_documents(documents) #This method splits your list of Document objects (documents you got from URLs) into smaller chunks. 
                                          #Each chunk is itself also a Document type
#Each Document will have two main parts: page_content → the actual text chunk (small piece of the webpage text).
#metadata → the information about where it came from (like URL, title, language, etc.).
```

```python
for i, chunk in enumerate(chunck):
    print(f"Chunk {i}:")
    print("Metadata:", chunk.metadata['source'])
    print("Page Content:", chunk.page_content)
    print("---------------------------------------------------")
```

```python
type(chunck) #chunks will be a list of Document objects.
print(chunck)
```

```python
chunck[10]
```

```python
from langchain_google_vertexai import VertexAIEmbeddings

embeddings = VertexAIEmbeddings(model="text-embedding-004")
```

```python
texts=[]
for chu in chunck:
    texts.append(chu.page_content)
print(texts)
```

```python
embed_result=embeddings.embed_documents(texts)
```

```python
from langchain_community.vectorstores import Chroma
```

```python
vector_store = Chroma.from_documents(
    documents=documents,# document objects
    embedding= embeddings # model
    #You don't need to first manually embed and pass embeddings.
#Chroma will call embeddings.embed_documents([...]) inside automatically.
#That’s why we pass model, not already embedded outputs.
)
```

```python
type(vector_store)
```

```python
retriver = vector_store.as_retriever()
```

```python
retriver_resultss = retriver.invoke("gen ai")
```

```python
print(retriver_resultss)
```

```python
for result in retriver_resultss:
    with open("web_rag.txt", "w") as file:
        file.write(result.page_content)
```

```python
from langchain.chat_models import init_chat_model
llm=init_chat_model("gemini-2.0-flash-001", model_provider="google_vertexai")
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt_quetion='what is gen ai'
promt_create_template=PromptTemplate.from_template('Answer the question based on the context below {retriver_resultss} and question is {prompt_quetion}')
```

```python
chain=promt_create_template | llm
```

```python
output=chain.invoke({'retriver_resultss':retriver_resultss,'prompt_quetion':prompt_quetion})
print(output.content)
```
